```python
#List of countries for the graph.
countries_to_plot = ['United States', 'Canada', 'Indonesia', 'Japan', 'Germany', 'Australia']
#Filtering the data frame
filtered_data = df[df['Country Name'].isin(countries_to_plot)]
sorted_data = filtered_data.sort_values(by='1997', ascending=False)
#Creating the bar chart
plt.figure(figsize=(12,4))
plt.bar (sorted_data['Country Code'], sorted_data['2019'],color='teal')
plt.xlabel ('Country Code')
plt.ylabel ('Population in 2019')
plt.title('Population of selected countries in 2019')
plt.xticks(rotation=45) #Rotating the x-axis labels for readability
plt.show()
```

```python
countries = df['Country Name']
population_2019 = df['2019']
#sorting the data
sorted_data = df.sort_values(by='2019', ascending=False)
#top 20 countries with highest population
top_countries = sorted_data['Country Name'].head(20)
plt.figure(figsize=(20, 5))
plt.bar(top_countries, sorted_data['2019'].head(20), color='skyblue')
plt.xlabel('Country Name')
plt.ylabel('Population in 2019')
plt.title(f'Distribution of Top 20 Populated Countries in 2019')
plt.xticks(rotation=45)  
plt.show()
```
